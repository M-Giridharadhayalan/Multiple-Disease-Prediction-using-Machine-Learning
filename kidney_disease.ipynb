```python
#there are so many null values in the dataset 
#Random Sampling for Higher Null Values
#Mean/Median for Lower Null Values
```

```python
def random_sampling(feature):
    random_sample = kdf[feature].dropna().sample(kdf[feature].isna().sum())
    random_sample.index=kdf[kdf[feature].isnull()].index
    kdf.loc[kdf[feature].isnull(),feature] = random_sample

def impute_mode(feature):
    mode = kdf[feature].mode()[0]
    kdf[feature]=kdf[feature].fillna(mode)
```

```python
#random sampling in numerical columns
for col in Numerical_columns:
    random_sampling(col)
```

```python
#categorical columns
random_sampling('red_blood_cells')
random_sampling('pus_cell')

for col in categorical_columns:
    impute_mode(col)
```

```python
#model building 
x=kdf.drop('class',axis=1)
y=kdf['class']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#KNN 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

knn = KNeighborsClassifier()
knn.fit(x_train,y_train)

knn_acc=accuracy_score(y_test,knn.predict(x_test))
print(f"Training Accuracy of KNN is {accuracy_score(y_train,knn.predict(x_train))}")
print(f"Testing Accuracy of KNN is {accuracy_score(y_test,knn.predict(x_test))}")


print(f"Confusion Matrix of KNN is \n {confusion_matrix(y_test,knn.predict(x_test))}")
print(f"Classification Report of KNN is \n {classification_report(y_test,knn.predict(x_test))}")
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
dtc_acc=accuracy_score(y_test,dtc.predict(x_test))
print(f"Training Accuracy of DTC is {accuracy_score(y_train,dtc.predict(x_train))}")
print(f"Testing Accuracy of DTC is {accuracy_score(y_test,dtc.predict(x_test))}")


print(f"Confusion Matrix of DTC is \n {confusion_matrix(y_test,dtc.predict(x_test))}")
print(f"Classification Report of DTC is \n {classification_report(y_test,dtc.predict(x_test))}")
```

```python
#Over Fitting of Training Result - Hyperparameter Tuning 
from sklearn.model_selection import GridSearchCV

GRID_PARAMETERS = {
    'criterion':['gini','entropy'],
    'max_depth':[3,5,7,10],
    'splitter':['best','random'],
    'min_samples_leaf':[1,2,3,5,7],
    'min_samples_split':[2,3,5,7],
    'max_features':['sqrt','log2']
}

grid_search_dtc=GridSearchCV(dtc,GRID_PARAMETERS,cv=5,n_jobs=-1,verbose=1)
grid_search_dtc.fit(x_train,y_train)
```

```python
#best parameter and best score 
print(grid_search_dtc.best_params_)
print(grid_search_dtc.best_score_)
```

```python
dtc=grid_search_dtc.best_estimator_

dtc_acc=accuracy_score(y_test,dtc.predict(x_test))
print(f"Training Accuracy of DTC is {accuracy_score(y_train,dtc.predict(x_train))}")
print(f"Testing Accuracy of DTC is {accuracy_score(y_test,dtc.predict(x_test))}")


print(f"Confusion Matrix of DTC is \n {confusion_matrix(y_test,dtc.predict(x_test))}")
print(f"Classification Report of DTC is \n {classification_report(y_test,dtc.predict(x_test))}")
```

```python
#random forest classifier 
from sklearn.ensemble import RandomForestClassifier
rf_class=RandomForestClassifier(criterion= 'gini', max_depth=7, max_features="log2", min_samples_leaf= 2, min_samples_split=5, n_estimators = 400)
rf_class.fit(x_train,y_train)
```

```python
rfc_score=accuracy_score(y_test,rf_class.predict(x_test))
print(f"Training Accuracy of RFC is {accuracy_score(y_train,rf_class.predict(x_train))}")
print(f"Testing Accuracy of RFC is {accuracy_score(y_test,rf_class.predict(x_test))}")


print(f"Confusion Matrix of RFC is \n {confusion_matrix(y_test,rf_class.predict(x_test))}")
print(f"Classification Report of RFC is \n {classification_report(y_test,rf_class.predict(x_test))}")
```

```python
#models data frame 
models = pd.DataFrame({ 'Model':['kneighbour','decision_trees','random_forest'],
                       'score':[knn_acc,dtc_acc,rfc_score]})
models.sort_values(by='score',ascending=False)
```
